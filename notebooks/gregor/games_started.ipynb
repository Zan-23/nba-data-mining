```python
def extract_games_played():
    games_played = []

    for year in range(0, 19):
    #for year in range(0, 19):

        file_str = ""

        if year < 9:
            file_str = "200{}-0{}".format(year, year + 1)
        elif year == 9:
            file_str = "200{}-{}".format(year, year + 1)
        else:
            file_str = "20{}-{}".format(year, year + 1)

        season_data = data_loader.load_data(seasons=[file_str])

        tmp = player_info[player_info["Season"] == file_str].apply(lambda x: season_data[((season_data["PLAYER1_ID"] == str(x["Player ID"])) | (season_data["PLAYER2_ID"] == str(x["Player ID"])) | (season_data["PLAYER3_ID"] == str(x["Player ID"])))]["GAME_ID"].nunique(), axis = 1)
        games_played.extend(tmp)
        
        print("{} ... done".format(file_str))
```
